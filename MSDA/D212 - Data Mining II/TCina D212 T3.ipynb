```python
#Part C1

import numpy as np
import pandas as pd
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori, association_rules

pd.set_option('display.max_columns', None)

csv = pd.read_csv("teleco_market_basket.csv", keep_default_na=False, header=0)
#print(csv.head(3))

#drop empty columns
csv["Item01"].replace('', np.nan, inplace=True)
csv.dropna(subset=["Item01"], inplace=True)

#transaction encode
transactions = csv.values.tolist()
encoder = TransactionEncoder().fit(transactions)
onehot = encoder.transform(transactions)

#save to dataframe, drop empty string column
onehot = pd.DataFrame(onehot, columns = encoder.columns_)
onehot.drop(onehot.columns[0], axis=1, inplace=True)

#save to file
onehot.to_csv("TCina D212 T3 Onehot.csv", index=False)
onehot.head(5)
```

```python
#Part C2

#compute frequent sets via apriori
frequent_itemsets = apriori(onehot, min_support = 0.0001, max_len = 2, use_colnames = True)

#compute association rules via lift
rules = association_rules(frequent_itemsets, metric = "lift", min_threshold = 1)

rules.sort_values(by="lift", ascending=False).head(10)
```
